```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data=pd.read_csv('nchs-birthrates-us.csv')
plt.figure(figsize=(10,6))
sns.lineplot(x='Year', y='Birth Rate', data=data)
plt.title('Birthrate Trends over Time')
plt.xlabel('Year')
plt.ylabel('Birth Rate')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=pd.read_csv('nchs-birthrates-us.csv')
plt.figure(figsize=(10,6))
x=data['Year']
y=data['Birth Rate']
plt.bar(x,y)
plt.title('Birthrate Trends over Time')
plt.xlabel('Year')
plt.ylabel('Birth Rate')
plt.show()
```
